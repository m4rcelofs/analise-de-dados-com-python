```python
from pandas import read_csv, to_datetime
from matplotlib import pyplot, style
```

```python
style.use("fivethirtyeight")
```

```python
df = read_csv("XRPBTC.csv")
df.set_index("Date", inplace=True)
df.index = to_datetime(df.index)
```

```python
df.corr()
```

```python
fig = pyplot.figure()
ax1 = pyplot.subplot2grid((1, 1), (0, 0))

df.plot(ax = ax1, label = "OHLCV data")
pyplot.show()
```

```python
ax1 = pyplot.subplot2grid((1, 1), (0, 0))

resampled_df = df["High"].resample("D").ohlc()
resampled_df.plot(ax = ax1, label = "OHLCV data")
pyplot.show()
```
